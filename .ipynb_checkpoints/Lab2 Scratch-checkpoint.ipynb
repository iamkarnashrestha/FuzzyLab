```python
A=dict()
def enter(name):
    list={}
    n=int(input("Enter the number of elements in set"+name))
    for i in range(n):
        name=input("Enter the name: ")
        while 1:
            value=float(input("Enter the value: "))
            if(value>=0 and value<=1):
                list[name]=value
                break;
            else:
                print("Value must be >= 0 and <=1")
    return list
def enteralpha():
    while 1:
        alpha=float(input("Enter the value of alpha"))
        if(alpha>=0 and alpha<=1):
            return alpha          
        else:
            print("Value must be >= 0 and <=1")
    
    
    
def Union(A,B):
    Y_set=dict()
    for A_key, B_key in zip(A, B):
        A_value = A[A_key]
        B_value = B[B_key]

        if A_value >= B_value :
            if A_value>0 or B_value>0:
                Y_set[A_key] = A_value
        elif B_value>A_value:
            Y_set[B_key] = B_value
          
    print('Union of crisp Set is :', Y_set)

    
def Intersection(A,B):
    Y_set=dict()
    for A_key, B_key in zip(A, B):
        A_value = A[A_key]
        B_value = B[B_key]
        
        if A_value <= B_value :
            if A_value>0 and B_value>0:
                Y_set[A_key] = A_value
        elif B_value<=A_value:
            if A_value>0 and B_value>0:
                Y_set[B_key] = B_value
    print('Intersection of crisp Set is :', Y_set)

def complement(A):
    Y_set=dict()
    for A_key in A:
        A_value = A[A_key]
        
        
        if 1-A_value !=0:
            Y_set[A_key] = 1-A_value
    return Y_set

def checksubset(A,B):
    countA=countB=0
    for A_key, B_key in zip(A,B):
        A_value = A[A_key]
        B_value = B[B_key]
    
        if B_value <= A_value :
            countA=countA+1
        if A_value<=B_value:
            countB=countB+1
        
    if(countA==len(B)):
        print("B is subset of A")
    if(countB==len(A)):
        print("A is subset of B")
    if((countA==len(B) or countB==len(A))!=True):
        print("Non are subset of each set")

def alphacut(A):
    alpha=enteralpha()
    Y_set=dict()
    for A_key in A:
        A_value = A[A_key]
        if alpha>=A_value and A_value>0:
            Y_set[A_key] = A_value
        
    print("The given set having alpha cut value",alpha,"is:",Y_set) 
    
def strictalphacut(A):
    alpha=enteralpha()
    Y_set=dict()
    for A_key in A:
        A_value = A[A_key]
        if alpha>A_value and A_value>0:
            Y_set[A_key] = A_value
        
    print("The given set having strict alpha cut value",alpha,"is:",Y_set) 
    
    
# A = {'a': 0.3, 'b': 0.6, 'd': 0.4}
# B= {'a': 0.1, 'b': 0.4, 'c': 0.2, 'e': 0.1}
A=enter("A")
B=enter("B")
print("Set A:",A)
print("Set B", B)
Union(A,B)
Intersection(A,B)
complement(A)
checksubset(A,B)
alphacut(A)
strictalphacut(A)
```
